## 2. Load Configurations & Data

```python
from utils.config import load_global_config
from utils.io_utils import load_parquet

config = load_global_config(PROJECT_ROOT)
symbol = config.get("data", "symbol", "BTCUSDT")

# Load features
features_path = os.path.join(PROJECT_ROOT, "data", "feature_store", f"{symbol}_5m_features.parquet")
df_features = load_parquet(features_path)

# Take a slice of the last 1000 candles for clear visualization
df_slice = df_features.iloc[-1000:].copy()
print(f"Slice shape for visualization: {df_slice.shape}")
```

## 3. Generate Interactive Candlestick & CVD Chart

```python
from visualization.plotly_charts import plot_candlestick_with_features

fig = plot_candlestick_with_features(df_slice, title=f"{symbol} Price Action, Sweeps, and CVD")
fig.show()
```

## 4. Save Dashboard to Google Drive

```python
dashboard_dir = os.path.join(PROJECT_ROOT, "reports", "dashboards")
os.makedirs(dashboard_dir, exist_ok=True)
dashboard_path = os.path.join(dashboard_dir, f"{symbol}_interactive_dashboard.html")

fig.write_html(dashboard_path)
print(f"Dashboard successfully saved as interactive HTML to {dashboard_path}")
```
